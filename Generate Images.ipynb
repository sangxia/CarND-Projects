```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.image as mpimg

%matplotlib inline
```

```python
pathname = 'data-track2/'
filename = pathname + 'driving_log.csv'
```

```python
df = pd.read_csv(filename, header=None, names=['center','left','right','steering','throttle','brake','speed'])
for col in ['center','left','right']:
    df.loc[:,col] = df[col].apply(lambda s: pathname+s[s.index('IMG'):])
```

```python
plt.rcParams["figure.figsize"] = (5,5)
plt.hist(df['steering'], bins=20)
```

```python
def plot_random(df, n_samples = 10):
    """
    plot n_samples random signs and their names
    """
    samples = np.random.choice(df.shape[0], size = n_samples, replace=False)
    n_per_row = 2
    nrows = int(np.ceil(n_samples/n_per_row))
    for i,idx in enumerate(samples):
        plt.subplot(nrows,n_per_row,i+1)
        plt.imshow(mpimg.imread(df.iloc[idx,0]))
        plt.title(df.iloc[idx,3])
    plt.tight_layout()
    plt.show()
```

```python
plt.rcParams["figure.figsize"] = (12,12)
plot_random(df, 10)
```
